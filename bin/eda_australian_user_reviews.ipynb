```python
# Tarea 1
# dado que las registros tienen valores nulos en las columnas 
# funny, posted, last_edited, item_id, helpful y recommend
# *** DECISIÓN: ELIMINAMOS las filas
filtro = df_DB[df_DB.item_id.isnull()].index
df_DB.drop(filtro, inplace=True)
```

```python
cant = len(df_DB.index)
print("Registros:", cant)
[(columna, df_DB[columna].isnull().sum()) for columna in columnas]
```

```python
df_DB.reset_index(inplace=True)
```

```python
# Tarea 2: crear la columna para año (anio_post)

# Remplazo carateres en el campo post para limpiar la fecha lo mejor posible
df_DB["posted"] = df_DB["posted"].str.replace("Posted ","").replace("posted ","").str.replace(".","").str.strip()
```

```python
# Creo la columna anio_post y lo lleno con valores del año tomados de la columna post
# a las fechas que estén en formato incorrecto se les coloca "0"
df_DB["anio_post"] = df_DB["posted"].apply(lambda x: int(x[len(x)-4:len(x)]) if x[len(x)-4:len(x)].isnumeric() else 0)
```

```python
# Reemplazo los valores de la columna anio_post que están en "0" por el año del 
# último registro anterior con año válido
print(anio:= df_DB[df_DB["anio_post"] > 0]["anio_post"].min())
for i in range(1,len(df_DB["anio_post"])):
    if int(df_DB["anio_post"][i]) == 0:
        df_DB["anio_post"][i] = anio
    else:
        anio = int(df_DB["anio_post"][i])
```

```python
# Tarea 3

# Creo variable con los nombres de la columas a eliminar
columnasEliminar = ["user_url", "funny", "posted", "last_edited", "helpful"]

# Elimino las columnas
df_DB.drop(columnasEliminar, axis=1, inplace=True)
```
